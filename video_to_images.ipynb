```python
import sys
import argparse
import os
import cv2
import numpy as np
print(cv2.__version__)
```

```python
def extractImages(pathIn, pathOut):
    count = 1
    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    success = True
    while success:
      vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*100))    # added this line 
      success,image = vidcap.read()
      print ('Read a new frame: ', success)
      ## Stop when last frame is identified
      image_last = cv2.imread("frame{}.png".format(count-1))
      if np.array_equal(image,image_last):
          break
      cv2.imwrite( os.path.join(pathOut,"frame{:d}.jpg".format(count)), image)     # save frame as JPEG file
      count = count + 1
```

```python
print("Taking images every 1/5th second")
pathIn = "dhl.mp4" 
pathOut = "H:\\projects\\logos32\\video" 
#H:\\projects\\logos32\\video
extractImages(pathIn, pathOut)
```
